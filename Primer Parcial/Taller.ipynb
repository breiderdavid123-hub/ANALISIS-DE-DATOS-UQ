Punto 1:
Para el total nacional calcule :
 
-población total
-población en edad de trabajar
-población que no está en edad de trabajar
-población económicamente activa
-población económicamente inactiva
-tasa de ocupación
-tasa de desempleo

```python
# Definir el Factor de Expansión
fex = 'FEX_C18'

# --- CÁLCULOS ---

# 1. Población Total
poblacion_total = df_cgsse[fex].sum()

# 2. Población en Edad de Trabajar (PET) -> Mayores de 15 años y menores de 65 años (P6040 es Edad)
pet = df_cgsse[(df_cgsse['P6040'] >= 15) & (df_cgsse['P6040'] <= 65)]['FEX_C18'].sum()

# 3. Población que no está en edad de trabajar
no_pet = poblacion_total - pet

# 4. Población Económicamente Activa (PEA) 
# Nota: La PEA está contenida en el archivo de Fuerza de Trabajo
pea = df_ft[(df_ft['FT'] == 1)]['FEX_C18'].sum()

# 5. Población Económicamente Inactiva (PEI)
# Nota: Son los "No ocupados" que no están buscando empleo activamente
pei = pet - pea

# 6. Ocupados y Desocupados para las tasas

oc = df_ocupados[(df_ocupados['OCI'] == 1)]['FEX_C18'].sum()

ds = df_desocupados[(df_desocupados['DSI'] == 1)]['FEX_C18'].sum()

# 7. Tasas
TO = (oc / pet) * 100
TD = (ds / pea) * 100

# --- RESULTADOS ---

print("--- INDICADORES MERCADO LABORAL (NACIONAL) ---")
print(f"Población Total: {poblacion_total:,.0f}")
print(f"Población en Edad de Trabajar (PET): {pet:,.0f}")
print(f"Población NO en Edad de Trabajar: {no_pet:,.0f}")
print(f"Población Económicamente Activa (PEA): {pea:,.0f}")
print(f"Población Económicamente Inactiva (PEI): {pei:,.0f}")
print(f"Tasa de Ocupación: {TO:.2f}%")
print(f"Tasa de Desempleo: {TD:.2f}%")
```

Explicacion Punto 1

aqui lo que se hizo fue sacar el calculo la población total sumando el factor de expansión de todos los registros del conjunto de datos. Posteriormente, se estima la población en edad de trabajar (PET) filtrando a las personas entre 15 y 65 años a partir de la variable de edad (P6040). A partir de esta información también se obtiene la población que no está en edad de trabajar como la diferencia entre la población total y la PET. Luego se calcula la población económicamente activa (PEA) utilizando el archivo de fuerza de trabajo, considerando las personas que participan en el mercado laboral. Con base en ello se determina la población económicamente inactiva (PEI) como la diferencia entre la PET y la PEA. Finalmente, se estiman los ocupados y desocupados mediante variables específicas de cada base de datos y, con estos valores, se calculan dos indicadores fundamentales del mercado laboral: la tasa de ocupación (TO), que mide la proporción de personas ocupadas dentro de la población en edad de trabajar, y la tasa de desempleo (TD), que representa la proporción de personas desocupadas dentro de la población económicamente activa.

Interpretacion de Resultados

A nivel nacional, Colombia tiene una población total de aproximadamente 52 millones de personas, de las cuales 35,3 millones están en edad de trabajar (15 a 65 años). La población económicamente activa alcanza casi 26 millones, con 23,7 millones ocupados y 2,2 millones desempleados. Esto se traduce en una tasa de ocupación del 67,26%, una tasa de desempleo del 8,55% y una tasa global de participación del 73,55%. Estos resultados muestran que, aunque la mayoría de las personas en edad de trabajar participa activamente en el mercado laboral, todavía existe un grupo significativo de desempleados, lo que indica la necesidad de políticas que fomenten la creación de empleo y la inclusión laboral.


Punto 2.

Calcule las tasas de desempleo para hombres y mujeres en Risaralda, Caldas , Tolima y Quindío. ¿Cuál es el departamento con mayor y menor tasa de desempleo de los 4 ? 


```python
# ======================================================
# CÁLCULO DE TASAS DE DESEMPLEO POR DEPARTAMENTO
# ======================================================

# Diccionario con los departamentos
departamentos = {
    63: "Quindío",
    66: "Risaralda",
    17: "Caldas",
    73: "Tolima"
}

# Diccionario para guardar las tasas
tasas = {}

# Recorremos cada departamento
for cod, nombre in departamentos.items():

    # Ocupados
    ocupados = df_ocupados[df_ocupados['DPTO'] == cod]['FEX_C18'].sum()

    # Desocupados
    desocupados = df_desocupados[df_desocupados['DPTO'] == cod]['FEX_C18'].sum()

    # PEA
    pea = ocupados + desocupados

    # Tasa de desempleo
    td = (desocupados / pea) * 100

    # Guardamos la tasa
    tasas[nombre] = td

# ======================================================
# RESULTADOS
# ======================================================

print("---- TASAS DE DESEMPLEO ----")

for dep, tasa in tasas.items():
    print(f"{dep}: {tasa:.2f}%")

# Departamento con mayor y menor tasa
mayor = max(tasas, key=tasas.get)
menor = min(tasas, key=tasas.get)

print("\n===================================")
print(f"Mayor tasa de desempleo: {mayor}")
print(f"Menor tasa de desempleo: {menor}")
print("===================================")
```

Explicacion Punto 2

En este punto se calcularon las tasas de desempleo para los departamentos de Quindío, Risaralda, Caldas y Tolima.Entonces primero se filtraron los registros de ocupados y desocupados según el código de cada departamento mediante la variable DPTO. Posteriormente, se sumó el factor de expansión FEX_C18 para obtener el total poblacional de ocupados y desocupados en cada territorio. Con estos valores se estimó la población económicamente activa (PEA) como la suma de ocupados y desocupados, y finalmente se calculó la tasa de desempleo dividiendo el número de desocupados entre la PEA y multiplicando el resultado por 100. Este procedimiento permitió comparar las tasas de desempleo entre los cuatro departamentos analizados.

Interpretacion de Resultados

Al analizar los cuatro departamentos, Quindío presenta la mayor tasa de desempleo general con 48,27%, mientras que Caldas muestra la más baja con 42,28%. Risaralda y Tolima tienen tasas intermedias de 44,75% y 46,45%, respectivamente. Estos datos evidencian que, aunque todos los departamentos presentan un nivel considerable de desempleo, existen diferencias regionales importantes, posiblemente relacionadas con la estructura económica local, las oportunidades de empleo disponibles y factores demográficos.

Punto 3

Tomando en consideración únicamente los siguientes grupos de edad: 

18 a 24,
24 a 30,
30 a 40,
40 a 50,
60 en adelante

Para Risaralda, Caldas , Tolima y Quindío, cual es el grupo de edad con mayor y menor tasa de desempleo (indicar en cada departamento el grupo de de edad  con mayor y menor tasa de desempleo) ?


```python
# =====================================================
# TASAS DE DESEMPLEO POR GRUPO DE EDAD Y DEPARTAMENTO
# =====================================================

# Función para crear los grupos de edad
def grupo_edad(edad):
    
    if 18 <= edad <= 24:
        return "18-24"
    
    elif 24 < edad <= 30:
        return "24-30"
    
    elif 30 < edad <= 40:
        return "30-40"
    
    elif 40 < edad <= 50:
        return "40-50"
    
    elif edad >= 60:
        return "60+"
    
    else:
        return None


# Crear variable grupo de edad
df_cgsse['grupo_edad'] = df_cgsse['P6040'].apply(grupo_edad)


# Crear identificador para unir bases
df_cgsse['ID'] = df_cgsse['DIRECTORIO'].astype(str) + "_" + df_cgsse['SECUENCIA_P'].astype(str) + "_" + df_cgsse['ORDEN'].astype(str)
df_ocupados['ID'] = df_ocupados['DIRECTORIO'].astype(str) + "_" + df_ocupados['SECUENCIA_P'].astype(str) + "_" + df_ocupados['ORDEN'].astype(str)
df_desocupados['ID'] = df_desocupados['DIRECTORIO'].astype(str) + "_" + df_desocupados['SECUENCIA_P'].astype(str) + "_" + df_desocupados['ORDEN'].astype(str)


# Traer edad y grupo de edad a ocupados y desocupados
edad = df_cgsse.set_index('ID')['grupo_edad']

df_ocupados['grupo_edad'] = df_ocupados['ID'].map(edad)
df_desocupados['grupo_edad'] = df_desocupados['ID'].map(edad)


# Departamentos
departamentos = {
    63: "Quindío",
    66: "Risaralda",
    17: "Caldas",
    73: "Tolima"
}

grupos = ["18-24","24-30","30-40","40-50","60+"]


# =====================================================
# CÁLCULO
# =====================================================

for dep, nombre in departamentos.items():
    
    print("\n--------------------------------")
    print("Departamento:", nombre)
    print("--------------------------------")
    
    tasas = {}
    
    for g in grupos:
        
        # Ocupados
        oc = df_ocupados[(df_ocupados['DPTO'] == dep) & 
                         (df_ocupados['grupo_edad'] == g)]['FEX_C18'].sum()
        
        # Desocupados
        ds = df_desocupados[(df_desocupados['DPTO'] == dep) & 
                            (df_desocupados['grupo_edad'] == g)]['FEX_C18'].sum()
        
        # Tasa de desempleo
        td = (ds / (oc + ds)) * 100 if (oc + ds) > 0 else 0
        
        tasas[g] = td
        
        print(f"{g}: {td:.2f}%")
    
    
    # Mayor y menor tasa
    mayor = max(tasas, key=tasas.get)
    menor = min(tasas, key=tasas.get)
    
    print("\nMayor tasa de desempleo:", mayor)
    print("Menor tasa de desempleo:", menor)
```

Explicacion Punto 3

En el tercer punto lo que se hizo fue identificar la edad de los individuos a partir de la variable correspondiente y se clasificaron en los grupos de edad establecidos: 18 a 24, 24 a 30, 30 a 40, 40 a 50 y 60 años o más. Posteriormente, se construyó un identificador único combinando las variables DIRECTORIO, SECUENCIA_P y ORDEN, lo que permitió relacionar correctamente la información entre las bases de datos de ocupados, desocupados y características generales. Con este identificador se incorporó el grupo de edad correspondiente a cada individuo dentro de las bases de ocupados y desocupados.

Una vez integrada la información, se procedió a filtrar los registros por departamento y grupo de edad, considerando únicamente los cuatro departamentos analizados. Luego se calcularon los totales de ocupados y desocupados en cada grupo mediante la suma del factor de expansión FEX_C18, el cual permite extrapolar los resultados de la muestra a la población total. Con estos valores se estimó la tasa de desempleo, aplicando la fórmula que divide el número de desocupados entre la suma de ocupados y desocupados y multiplicando el resultado por cien. Finalmente, se compararon las tasas obtenidas dentro de cada departamento para identificar cuál grupo de edad presenta la mayor y la menor tasa de desempleo, lo que permite observar cómo varía la situación del mercado laboral según el ciclo de vida de la población.

Intepretacion de los resultados

El análisis por edades revela que las personas de 60 años en adelante enfrentan las mayores dificultades para conseguir empleo, con tasas de desempleo superiores al 70% en todos los departamentos. Por otro lado, los jóvenes entre 25 y 30 años presentan las tasas más bajas, alrededor del 21–32%, dependiendo del departamento. Esto sugiere que la experiencia laboral y la flexibilidad laboral pueden favorecer la inserción de los jóvenes, mientras que los adultos mayores se enfrentan a mayores barreras de acceso al empleo, lo que resalta la necesidad de programas de empleo dirigidos a este grupo etario.


Punto 4

Entre Risaralda, Caldas , Tolima y Quindío, cuál es el departamento con menor tasa de desempleo de mujeres entre los 18 y los 25 años ?

```python
# ======================================================
# CREAR IDENTIFICADOR ÚNICO
# ======================================================

df_cgsse['ID_TEMP'] = (
    df_cgsse['DIRECTORIO'].astype(str) +
    df_cgsse['SECUENCIA_P'].astype(str) +
    df_cgsse['ORDEN'].astype(str)
)

df_ocupados['ID_TEMP'] = (
    df_ocupados['DIRECTORIO'].astype(str) +
    df_ocupados['SECUENCIA_P'].astype(str) +
    df_ocupados['ORDEN'].astype(str)
)

df_desocupados['ID_TEMP'] = (
    df_desocupados['DIRECTORIO'].astype(str) +
    df_desocupados['SECUENCIA_P'].astype(str) +
    df_desocupados['ORDEN'].astype(str)
)

# TRAER SEXO Y EDAD DESDE CGSSE

dic_sexo = df_cgsse.set_index('ID_TEMP')['P3271']
dic_edad = df_cgsse.set_index('ID_TEMP')['P6040']

df_ocupados['SEXO_EXACTO'] = df_ocupados['ID_TEMP'].map(dic_sexo)
df_desocupados['SEXO_EXACTO'] = df_desocupados['ID_TEMP'].map(dic_sexo)

df_ocupados['EDAD_EXACTA'] = df_ocupados['ID_TEMP'].map(dic_edad)
df_desocupados['EDAD_EXACTA'] = df_desocupados['ID_TEMP'].map(dic_edad)

print("DESEMPLEO MUJERES ENTRE 18 Y 25 AÑOS")

# QUINDÍO
oc_qui = df_ocupados[(df_ocupados['DPTO']==63) &
                     (df_ocupados['SEXO_EXACTO']==2) &
                     (df_ocupados['EDAD_EXACTA']>=18) &
                     (df_ocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

ds_qui = df_desocupados[(df_desocupados['DPTO']==63) &
                        (df_desocupados['SEXO_EXACTO']==2) &
                        (df_desocupados['EDAD_EXACTA']>=18) &
                        (df_desocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

td_qui = (ds_qui/(oc_qui+ds_qui))*100


# RISARALDA
oc_ris = df_ocupados[(df_ocupados['DPTO']==66) &
                     (df_ocupados['SEXO_EXACTO']==2) &
                     (df_ocupados['EDAD_EXACTA']>=18) &
                     (df_ocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

ds_ris = df_desocupados[(df_desocupados['DPTO']==66) &
                        (df_desocupados['SEXO_EXACTO']==2) &
                        (df_desocupados['EDAD_EXACTA']>=18) &
                        (df_desocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

td_ris = (ds_ris/(oc_ris+ds_ris))*100


# CALDAS
oc_cal = df_ocupados[(df_ocupados['DPTO']==17) &
                     (df_ocupados['SEXO_EXACTO']==2) &
                     (df_ocupados['EDAD_EXACTA']>=18) &
                     (df_ocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

ds_cal = df_desocupados[(df_desocupados['DPTO']==17) &
                        (df_desocupados['SEXO_EXACTO']==2) &
                        (df_desocupados['EDAD_EXACTA']>=18) &
                        (df_desocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

td_cal = (ds_cal/(oc_cal+ds_cal))*100


# TOLIMA
oc_tol = df_ocupados[(df_ocupados['DPTO']==73) &
                     (df_ocupados['SEXO_EXACTO']==2) &
                     (df_ocupados['EDAD_EXACTA']>=18) &
                     (df_ocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

ds_tol = df_desocupados[(df_desocupados['DPTO']==73) &
                        (df_desocupados['SEXO_EXACTO']==2) &
                        (df_desocupados['EDAD_EXACTA']>=18) &
                        (df_desocupados['EDAD_EXACTA']<=25)]['FEX_C18'].sum()

td_tol = (ds_tol/(oc_tol+ds_tol))*100


print("Quindío:", round(td_qui,2),"%")
print("Risaralda:", round(td_ris,2),"%")
print("Caldas:", round(td_cal,2),"%")
print("Tolima:", round(td_tol,2),"%")


tasas = {
    "Quindío": td_qui,
    "Risaralda": td_ris,
    "Caldas": td_cal,
    "Tolima": td_tol
}

menor = min(tasas, key=tasas.get)

print("\nEl departamento con menor tasa de desempleo es:", menor)
```

Punto 7

¿ Qué porcentaje de la población de Risaralda es población económicamente inactiva ?

```python
# =====================================================================
# POBLACIÓN ECONÓMICAMENTE INACTIVA (PEI) EN RISARALDA
# =====================================================================

print("--- POBLACIÓN ECONÓMICAMENTE INACTIVA (PEI) EN RISARALDA ---")

# 1. Calculamos la población TOTAL de Risaralda (DPTO == 66)
pt_ris = df_cgsse[df_cgsse['DPTO'] == 66]['FEX_C18'].sum()

# 2. Calculamos la PET de Risaralda (15 a 65 años)
pet_ris = df_cgsse[(df_cgsse['DPTO'] == 66) & 
                   (df_cgsse['P6040'] >= 15) & 
                   (df_cgsse['P6040'] <= 65)]['FEX_C18'].sum()

# 3. Calculamos la PEA de Risaralda (Desde el módulo Fuerza de Trabajo)
pea_ris = df_ft[(df_ft['DPTO'] == 66) & (df_ft['FT'] == 1)]['FEX_C18'].sum()

# 4. Los Inactivos (PEI) se sacan por diferencia (PET - PEA)
pei_ris = pet_ris - pea_ris

# 5. Calculamos el porcentaje de PEI sobre la población total
porc_pei_total = (pei_ris / pt_ris) * 100

# 6. Imprimimos el resultado
print(f"Población Total de Risaralda: {pt_ris:,.0f} personas")
print(f"Población Económicamente Inactiva (PEI): {pei_ris:,.0f} personas\n")

print(f"RESPUESTA: El {porc_pei_total:.2f}% de la población de Risaralda es población económicamente inactiva.")
```

Explicacion Punto 7

En este punto se calculó el porcentaje de la población económicamente inactiva (PEI) en el departamento de Risaralda. Primero se obtuvo la población total del departamento utilizando la base de características generales (df_cgsse) y sumando el factor de expansión (FEX_C18) para las observaciones con DPTO = 66. Posteriormente se calculó la población en edad de trabajar (PET) filtrando las personas entre 15 y 65 años según la variable de edad (P6040). Luego se estimó la población económicamente activa (PEA) utilizando el módulo de fuerza de trabajo (df_ft), filtrando las personas cuya variable FT es igual a 1. Finalmente, la población económicamente inactiva se obtuvo restando la PEA a la PET, y se calculó el porcentaje que representa esta población respecto a la población total del departamento.

Interpretacion de los Resultados

En el Tolima los jóvenes ocupados entre 20 y 25 años representan un porcentaje específico de la población total del departamento, lo que permite medir su inserción laboral relativa. Estos datos muestran que, aunque un número importante de jóvenes se encuentra trabajando, todavía existe un segmento que no logra integrarse al mercado laboral, evidenciando la necesidad de estrategias de inclusión y fortalecimiento de habilidades.
